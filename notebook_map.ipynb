```python
Coordinates = namedtuple("Coordinates", "lat lon")
moscow_coords = Coordinates(55.75222, 37.61556)
ekaterenburg_coords = Coordinates(56.838008, 60.597452)
```

```python
with open("data_1.json") as f:
    request = json.load(f)
```

```python
scooter_number = len(request["points"])
center_mass = Coordinates(
    sum([p["position"][0] for p in request["points"]]) / scooter_number,
    sum([p["position"][1] for p in request["points"]]) / scooter_number,
)
delta = Coordinates(
    ekaterenburg_coords.lat - center_mass.lat, ekaterenburg_coords.lon - center_mass.lon
)
```

```python
map = folium.Map(location=[*ekaterenburg_coords], zoom_start=11)

for point in request["points"]:
    lat, lon = point["position"]

    folium.Marker(
        location=[lat + delta.lat, lon + delta.lon],
        popup=point["position"],
        icon=folium.Icon(icon="star", color="orange"),
    ).add_to(map)
map.save("map.html")
```
